# Comandos básicos de Pandas

aqui veremos las funcionalidades de pandas en los entornos parecidos a tablas o datos ordenados los cuales podremos saber la informacion de los mismos e incluso hacerless transformaciones por medio de consultas y filtros para usarlo simplemente hay que ir dandole play a cada uno de las partes del codigo

```python
import pandas as pd

data = {
    'Nombre': ['Ana', 'Luis', 'Marta', 'Carlos', 'Lucía'],
    'Edad': [23, 35, 29, 42, 31],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia', 'Bilbao'],
    'Salario': [2500, 3000, 2800, 3500, 2700]
}

df = pd.DataFrame(data)
df
```

## Información general del DataFrame

```python
df.info()
```

## Primeras y últimas filas

```python
df.head()
```

```python
df.tail()
```

## Selección de columnas y filas

```python
df['Nombre']
```

```python
df[['Nombre', 'Edad']]
```

```python
df.iloc[0:3]
```

```python
df.loc[0:2, ['Nombre', 'Ciudad']]
```

## Filtrado de datos

```python
df[df['Edad'] > 30]
```

## Ordenar valores

```python
df.sort_values(by='Edad')
```

```python
df.sort_values(by='Salario', ascending=False)
```

## Estadísticas básicas

```python
df.describe()
```

## Agrupaciones

```python
df.groupby('Ciudad')['Salario'].mean()
```

## Guardar DataFrame en CSV

```python
df.to_csv('datos.csv', index=False)
```
