# 📊 Comandos Polars

## Introducción

Polars es una librería de procesamiento de datos rápida escrita en Rust y optimizada para análisis de datos en Python. Es una alternativa moderna a pandas, diseñada para ser más eficiente en memoria y velocidad.

### ¿Qué se puede hacer con Polars?
- Manipulación de datos (DataFrames)
- Limpieza y transformación de datos
- Análisis de grandes volúmenes de información
- Operaciones rápidas con múltiples núcleos
- Procesamiento lazy (optimización automática)

````python
import polars as pl
````

## Crear DataFrames

````python
df = pl.DataFrame({
    "nombre": ["Ana", "Luis", "Pedro"],
    "edad": [23, 35, 29]
})
df
````

## Lectura de archivos

````python
df_csv = pl.read_csv("archivo.csv")
df_parquet = pl.read_parquet("archivo.parquet")
````

## Selección y filtrado

````python
df.select(["nombre"])
df.filter(pl.col("edad") > 25)
````

## Agregaciones

````python
df.select([
    pl.col("edad").mean(),
    pl.col("edad").max()
])
````

## GroupBy

````python
df.groupby("nombre").agg([
    pl.col("edad").sum()
])
````

## Columnas nuevas

````python
df.with_columns([
    (pl.col("edad") * 2).alias("edad_doble")
])
````

## Ordenar

````python
df.sort("edad", descending=True)
````

## Joins

````python
df2 = pl.DataFrame({
    "nombre": ["Ana", "Luis"],
    "ciudad": ["Bogotá", "Medellín"]
})

df.join(df2, on="nombre", how="inner")
````

## Lazy API (optimización)

````python
lazy_df = df.lazy()
lazy_df.filter(pl.col("edad") > 25).collect()
````

## Exportar datos

````python
df.write_csv("salida.csv")
````
