# Ejemplos de uso de Matrices con NumPy

las matrice son muy importantes para el analisis de datos ya que de esta forma se puede representar en que parte puede estar guardada cierta informacion y como poder buscarla por medio de nunpy tambien hay otros archivos los cuales describen como buscar o filtrar o transformar la informacion por medio de filtros o ecuaciones matematicas o incluso con cambios de datos 

```python
import numpy as np
```

## Ejemplo 1: Crear una matriz

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
A
```

## Ejemplo 2: Operaciones básicas

```python
B = np.array([[7, 8, 9], [10, 11, 12]])

suma = A + B
resta = B - A
producto = A * B  # elemento a elemento

suma, resta, producto
```

## Ejemplo 3: Producto matricial

```python
C = np.array([[1, 2], [3, 4], [5, 6]])

producto_matriz = np.dot(A, C)
producto_matriz
```

## Ejemplo 4: Transpuesta e inversa

```python
M = np.array([[1, 2], [3, 4]])

transpuesta = M.T
inversa = np.linalg.inv(M)

transpuesta, inversa
```

## Ejemplo 5: Determinante y valores propios

```python
det = np.linalg.det(M)
valores, vectores = np.linalg.eig(M)

det, valores, vectores
```
