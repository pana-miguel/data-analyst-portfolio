Se implementa un sistema interactivo en consola que permite filtrar información según las cedulas o la edad para saber cuantes ventas de vehiculos se han vendido mas, 

El uso de Polars en este caso ofrece un alto rendimiento al trabajar con archivos grandes, optimizando tanto el tiempo de carga como las operaciones de filtrado ya que el archivo principal tiene 1 millon de filas para anlizar como se muestra en la imagen el resultado de una prueba y el rendimiento que tubo polars a la hora de ejcutar esta funcion.
 ![Texto alternativo](../../../../otros-elementos/venta-vehiculos-resultado-funciones-polars.png)

 de esta forma es como se tiene que ver el resultado de la prueba

````python
%%time
import polars as pl

# Cargar CSV
archivo = pl.read_csv("ventas_vehiculos_simuladas1.csv")
print(f"Total de registros cargados: {len(archivo):,}\n")

# Funciones de búsqueda
def buscar_por_cedula(df):
    valor = input("Ingresa la cédula: ").strip()
    return df.filter(pl.col("cedula").cast(pl.Utf8) == valor)

def buscar_por_edad(df):
    valor = int(input("Ingresa la edad: ").strip())
    return df.filter(pl.col("edad") == valor)

def buscar_por_anio(df):
    anio = int(input("Ingresa el año (ejemplo: 2023): ").strip())
    df = df.with_columns(pl.col("fecha_venta").str.strptime(pl.Date, "%Y-%m-%d"))
    return df.filter(pl.col("fecha_venta").dt.year() == anio)

def buscar_por_celular(df):
    valor = int(input("Ingresa el número celular: ").strip())
    return df.filter(pl.col("numero_celular") == valor)

def buscar_por_valor_vehiculo(df):
    valor = float(input("Ingresa el valor del vehículo: ").strip())
    return df.filter(pl.col("valor_vehiculo") == valor)

def buscar_por_iva(df):
    valor = int(input("Ingresa el valor del IVA: ").strip())
    return df.filter(pl.col("iva") == valor)

def buscar_por_valor_total(df):
    valor = float(input("Ingresa el valor total: ").strip())
    return df.filter(pl.col("valor_total") == valor)


# “Switch” con funciones mapeadas
busquedas = {
    "1": buscar_por_cedula,
    "2": buscar_por_edad,
    "3": buscar_por_anio,
    "4": buscar_por_celular,
    "5": buscar_por_valor_vehiculo,
    "6": buscar_por_iva,
    "7": buscar_por_valor_total
}

# Menú dinámico
print("=== FORMULARIO DE BÚSQUEDA ===")
print("1. Buscar por cédula")
print("2. Buscar por edad")
print("3. Buscar por año de venta")
print("4. Buscar por número celular")
print("5. Buscar por valor del vehículo")
print("6. Buscar por IVA")
print("7. Buscar por valor total\n")

opcion = input("Elige el tipo de búsqueda (1–7): ").strip()

# Ejecución del "switch"
funcion = busquedas.get(opcion)

if funcion:
    resultado = funcion(archivo)
    if resultado.height > 0:
        print(f"Se encontraron {resultado.height} coincidencias.\n")
        print(resultado)
    else:
        print("No se encontraron resultados.")
else:
    print("Opción no válida. Saliendo...")
````
