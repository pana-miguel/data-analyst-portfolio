#### En esta parte lo que hacemos es utilizar polars para que recople todos los datos ya que este trabaja mejor que pandas cuando sse trata de millones de datos despues tomamos las cedulas simuladas 1 y 2 y las transformamos con el metodo concat que es concatenar las dos en un solo archivo para finalmente exportarlo y tambien posteriormente a las cedulas se les hace un cambio de tipo de dato ya que lo nesecitaremos de esta forma para que nuestro algoridmo generico que esta en [analisis.ipyn](./analisis.ipynb) se pueda usar de una forma correcta y muy parecida al funcionamiento interno de polars 

```python
import polars as pl

# Lista de archivos a unir
archivos = [
    "cedulas-simuladas1.csv",
    "cedulas-simuladas2.csv",
]
# Leer y unir todos
df_total = pl.concat([pl.read_csv(a) for a in archivos])

# Guardar el archivo combinado
df_total.write_csv("cedulas-simuladas-total1.csv")
```

#### Cambio de tipo de dato de las cedulas a utf8 para que no las lea como numeros sino como un texto 

```python
import polars as pl

archivo = pl.read_csv("cedulas-simuladas-total1.csv")

# Convertir la columna 'cedula' a texto (Utf8) dentro del DataFrame
archivo = archivo.with_columns(
    pl.col("cedula").cast(pl.Utf8)
)
# Verificar que el cambio se aplicó
print(list(zip(archivo.columns, archivo.dtypes)))

archivo.write_csv("cedulas-simuladas-total1.csv")
```
