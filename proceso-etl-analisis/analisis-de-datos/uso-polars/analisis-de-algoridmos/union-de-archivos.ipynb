```python
import pandas as pd

# Lista de archivos a unir
archivos = [
    "cedulas_simuladas_1.csv",
    "cedulas_simuladas_2.csv",
    "cedulas_simuladas_3.csv",
    "cedulas_simuladas_4.csv"
]

# Leer y unir todos
df_total = pd.concat([pd.read_csv(a) for a in archivos])

# Guardar el archivo combinado
df_total.to_csv("cedulas_simuladas_total.csv", index=False)

```

```python
import polars as pl

archivo = pl.read_csv("cedulas_simuladas_total.csv")

# Convertir la columna 'cedula' a texto (Utf8) dentro del DataFrame
archivo = archivo.with_columns(
    pl.col("cedula").cast(pl.Utf8)
)

# Verificar que el cambio se aplicó
print(list(zip(archivo.columns, archivo.dtypes)))

archivo.write_csv("cedulas_simuladas_total.csv")
```
